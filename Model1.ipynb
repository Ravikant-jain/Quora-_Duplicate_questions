# Simple BOW with RandomForest

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_df=pd.read_csv(r'D:\Github\NLP_Duplicate_questions\quora-question-pairs\train.csv\train.csv')
all_df
```

```python
df=all_df.sample(50000)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df=df.dropna()
df.isna().sum()
```

```python
ques_df=df[['question1','question2']]
ques_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
all_ques=list(ques_df['question1']) + list(ques_df['question2'])
len(all_ques)
```

```python
cv=CountVectorizer(max_features=3000)
q1,q2=np.vsplit(cv.fit_transform(all_ques).toarray(),2)
```

```python
tdf1=pd.DataFrame(q1,index=ques_df.index)
tdf2=pd.DataFrame(q2,index=ques_df.index)
tdf2.shape
```

```python
temp_df=pd.concat([tdf1,tdf2],axis=1)
temp_df.shape
```

```python
temp_df.head()
```

```python
temp_df['is_duplicate']=df['is_duplicate']
```

```python
temp_df
```

```python
temp_df.to_csv('BOW_Data.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(temp_df.iloc[:,0:-1].values,temp_df.iloc[:,-1].values,test_size=0.2,random_state=69)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```
